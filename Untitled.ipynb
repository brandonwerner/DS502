```python
def filter_county(df):

# a function that removes " county" from the county name, so names match
# apply to each state mortality rate file 

    for i in df.index:
        old_name = df.iloc[i]["location_name"]
    
        end = len(old_name)

        delete = len(" County")
    
        new_name = old_name[0:end-delete]
        df.set_value(i, "location_name", new_name)
    
    
    return df
```
